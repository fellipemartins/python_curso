<a name="oque_variaveis">
    <H1>O que são variáveis?</H1>
</a>

A ideia básica de programação é manipular (fazer todo tipo de operação) em valores (números, texto, etc). E o componente básico dessa manipulação são as **variáveis**.

Vamos começar com uma analogia - imagine que você trabalha numa loja de peças automotivas gerenciando o estoque de peças. As peças estão num depósito no fundo da loja empacotadas em caixas. Há diversos tipos e tamanhos de caixas, e cada caixa pode caber uma ou mais peças - inclusive de tipos e tamanhos diferentes. Aqui temos os dois elementos principais: o *conteúdo* que vai em cada caixa (isto é, os **valores**) e cada *formato* de caixa com suas características (os **tipos**).

> **Exemplo**: em `Python`, você pode criar uma variável ao definir um *nome* e associar a um *valor*: `ano = 365`

Neste exemplo vimos claramente o *valor* (365) e o *nome* da variável (ano). Onde então está o tipo?

Essa noção é importante porque há linguagens de programação que são *explicitamente tipadas* (C++ por exemplo) e outras que são *dinamicamente tipadas* como `Python` e `R` - `Julia` por outro lado é *opcionalmente* tipada.

O que isto quer dizer é que em algumas linguagens de programação você precisa deixar claro o *tipo* de *valores* podem ser contidos na variável durante sua declaração (i.e., `tipo nome = valor`). Veja um exemplo em C++:

> **Exemplo**: `int ano = 365;`

Isto quer dizer que no paradigma explícito você precisa informar três coisas - o nome da variável (*ano*), o tipo que ela recebe (números inteiros `int`) e seu valor (ou *conteúdo* - 365). Isto é, o tipo foi *explicitamente definido* durante a declaração da variável (daí a expressão "explicitamente tipada").

`Python`, por outro lado, é *dinamicamente* tipada. Portanto, não é necessário informar ao interpretador que tipo de dados contém cada variável. Como funciona então? Nos bastidores, o próprio `Python` "lê" o conteúdo e infere seu tipo).

<br>

<img src="data/tipagem2.png" width=750>

> Exemplo: Se você mudar em C++, `ano = "2021";`, retornará um erro, porque `ano` só recebe números inteiros, e não texto (valores entre aspas, por exemplo). Em `Python`, por outro lado, é perfeitamente aceitável (e também perigoso em certas situações) mudar de `ano = 365` para `ano = "2021"`.

A vantagem da abordagem dinâmica é que requer menos trabalho e planejamento por parte do programador. Mas há desvantagens nesse tipo de abordagem: 

* Faz com que os programas se tornem mais lentos (é um passo a mais para o interpretador executar). 
* Quando você declara o tipo da variável previamente, você faz uma "reserva" de memória. Em casos mais adiante vamos ver que isso faz com que a "reserva" seja maior do que o necessário, fazendo com que um conjunto de dados ocupe mais espaço que o necessário. 
* Um problema adicional é que pode haver erros na interpretação do tipo de dado levando a problemas de análise de dados feita de forma incorreta

Voltando à analogia da loja de autopeças. Podemos delimitar uma série de variáveis sobre uma peça:

- Nome
- Marca
- Peso
- Preço
- Código
- Quantidade

Perceba que cada variável a respeito de uma peça tem características intrínsecas: "nome" e "marca" serão texto, "quantidade" será um número inteiro (5 ou 6 peças, mas não 3,7 peças), "peso" e "preço" serão fracionados (8,9 kg; R$ 240,76) e assim por diante. 

Cada linguagem de programação difere em como tratar os tipos de variáveis. Vamos ver em seguida como `Python` lida com esse assunto.

> **Dica**: Se você planeja coletar dados em formato tabular, cada coluna terá um nome semelhante à variável. Crie um arquivo separado chamado de "dicionário de variáveis". Escreva o nome da variável como se encontra no quadro de dados e sua definição - o que é, qual a medida, detalhes, etc. Isso facilita o trabalho de quem for futuramente rever seus dados (inclusive você mesmo depois de muito tempo)

    <H1>Declarando variáveis</H1>
</a>

Vimos um exemplo prévio de declaração de variáveis em `Python`, mas há alguns detalhes importantes. Primeiro, `Python` é sensível a tipos de letras (*case sensitive*) - isto quer dizer que `ano`, `Ano`, `ANO` e `AnO` serão interpretados como coisas diferentes. 

Segundo, a ordem importa - primeiro vem o nome, o sinal de `= ` (que indica "recebe") e o valor pretendido.

> **Dica**: O símbolo de `=` só é usado para atrelar valores a variáveis e por isso `=` NÃO é usado para teste de igualdade. Mais adiante você verá que para fazer testes de igualdade `Python` usa `==`.

Em `Python` você terá um erro de sintaxe se tentar fazer o contrário (`365 = ano`).

````python
# Leia: ano recebe 365
ano = 365
````

````python
# erro
# 365 = ano
````

Compare com `R` em que você pode fazer em ambas as direções ( `ano <- 365`, `365 -> ano`, `ano = 365` ou `365 = ano`).

`Python` tem um recurso chamado *unpacking* (desempacotamento) que serve para declarar mais de uma variável ao mesmo tempo (contanto que a quantidade de valores e de variáveis seja igual). Uma desvantagem de fazer isso é na hora de alterar o código, legibilidade e organização. 

Exemplo:

````python
ano, mes, dia, ocasiao, venda = 365, 12, 25, "natal", 55.89
````

### Nomeando variáveis

`Python` limita a forma como você nomeia uma variável. Primeiro, você não pode nomear uma variável com umas das palavras protegidas da linguagem (*keywords*). 

Vamos verificar as palavras protegidas do `Python`:

````python
import keyword
print(keyword.kwlist)
````

Como exemplo, vamos criar uma variável com um texto igual a uma das palavras protegidas, e vamos testar se está na lista:

````python
nome_da_variavel = "while"
nome_da_variavel in keyword.kwlist
````

````python
# erro
# nome_da_variavel_2 = while
````

Segundo, há outras restrições de nomeação de variáveis:

* Pode ter qualquer tamanho (mas use seu bom senso)
* Pode conter letras maiúsculas ("A, B, ...") e minúsculas ("a, b, ...")
* Pode conter *underline* (`_`)
* Pode conter dígitos (0, 1, 2...) - mas **não pode iniciar com dígitos**

Nomes de variáveis válidos: `nomeDoCliente`, `mêsdonascimento`, `receita2022`, etc.


No entanto, a equipe mantenedora do `Python` edita regularmente guias da linguagem e um desses ([`PEP-8`](https://www.python.org/dev/peps/pep-0008/)) define algumas regras que ajudam na homogeneização da escrita em `Python`. Sugiro que você leia essas regras depois, já que há regra para cada tipo de coisa (variável, função, classe, etc).

Para variáveis, sugere-se a convenção *snake case*:

* Usar letras minúsculas
* Separar as palavras (caso haja mais de uma) com um `_`

Assim, os nomes de variáveis acima ficariam: `nome_do_cliente`, `mês_do_nascimento`, `receita_2022`,

Outras sugestões são retirar palavras como "do", "da", "para", etc., entre palavras de uma variável e criar variáveis sem acento ou sinais (o que pode dar erro de codificação dependendo do caso) - portanto  `mês_do_nascimento` ficaria melhor como  `mes_nascimento`.

> **Dica**: A exceção são as constantes  - isto é, valores que você não pretende alterar durante a execução do código. Sugere-se que as constantes sejam em letras maiúsculas, portanto `PI = 3.14159`.

> **Dica**: Quando for preparar dados tabulares ou planejar uma coleta de dados, nomeie as colunas com essas convenções acima. Isso facilitará muito o trabalho de limpeza nos nomes futuramente.

<a name="booleanos"> <H1>Booleanos (valores do tipo lógico)</H1> </a>

`Python`, como outras linguagens de programação, divide o mundo entre ***verdadeiro*** e ***falso***. Assim, diversas operações podem ser feitas para testar se algo é verdadeiro ou falso.

### Estilo `str.format()`

No estilo `str.format()`, você cria um texto como antes, usando aspas simples ou duplas. O pulo do gato é passar um método chamado `.format()` que permite você manipular as variáveis que aparecem no texto, por fora. Como se faz? O lugar onde a variável deve ficar no texto é substituída por `{}`, e a referência à variável é passada como parâmetro do método `.format()`:

````python
# primeira versão
"Um ano tem", ano, "dias!"

# segunda versão com str.format()
"Um ano tem {} dias!".format(ano)
````

Se houver mais de uma variável, basta com que a ordem seja obedecida para o interpretador saber qual par de colchetes equivale a qual variável:

````python
"Um ano tem {} dias, e cada mês em média tem {}!".format(ano, ano / 12)
````

À primeira vista pode parecer que dá mais trabalho, mas é mais reproduzível e mais fácil de mudar depois. Uma vantagem de usar `.format()` é que você pode referenciar as variáveis fora de ordem, indicando dentro dos `{}` seu nome:

````python
mes = ano / 12

"Um ano tem {ano} dias, e cada mês em média tem {mes}!".format(mes=mes, ano=ano)
````

Você pode definir as variáveis usando `.format()` de forma separada também, que dependendo do caso pode ser útil (veremos `if`, `for`, etc. mais adiante):

````python
local = input("Qual a sua localização? SP / RJ / MG").upper()

preco = 0
preco_SP = 45.56
preco_RJ = 40.99
preco_MG = 39.99

if local == "SP":
    preco = preco_SP
elif local == "MG":
    preco = preco_MG
else:
    preco = preco_RJ

                            # marcador de posição "preco"
aviso = "O produto custa R$ {preco}, compre antes que acabe!"
                   # marcador de posição --> variável
print(aviso.format(preco = preco))
````

### Estilo `f"string"`

O estidlo `f"string"` é similar ao anterior (`str.format()`). A grande diferença é que você já informa ao interpretador que o texto tem alguma formatação prefixando o texto com `f" .... "` e passa a variável como argumento dentro dos `{}`:

````python
# primeira versão intercalada
"Um ano tem", ano, "dias!"

# segunda versão com str.format()
"Um ano tem {} dias!".format(ano)

# terceira versão com f-string:
f"Um ano tem {ano} dias!"
````

Da mesma forma, você pode fazer operações dentro dos `{}`:

````python
# segunda versão com str.format()
"Um ano tem {} dias, e cada mês em média tem {}!".format(ano, ano / 12)

# terceira versão com f-string:
f"Um ano tem {ano} dias, e cada mês em média tem {ano / 12} dias!"
````

Você pode formatar como uma variável é apresentada entre colchetes. Um exemplo é o caso acima, em que fizemos a operação `ano / 12` - o resultado é preciso, mas apresenta muitas casas decimais. Você até poderia usar a função nativa `round()` para arredondar, mas há uma forma interessante de fazer isso direto nos colchetes: `{variável:.Xf}` em que `X` é o número de casas decimais a apresentar. Exemplo:

````python
# sem delimitar casas:
print(f"Um ano tem {ano} dias, e cada mês em média tem {ano / 12} dias!")

# delimitando casos:
print(f"Um ano tem {ano} dias, e cada mês em média tem {ano / 12:.1f} dias!")
````

### Entrada de texto

Apesar de provavelmente não usarmos mais adiante, caso você tenha ficado curioso sobre entrada de texto, geralmente se faz com a função `input()` atrelada a uma variável. Você pode passar um texto dentro dos parênteses como mensagem:

````python
nome = input("Qual é o seu nome?")
print(f"Seu nome é {nome} então. Blz!")
````

A função `input()` **sempre** guarda a entrada como texto:

````python
anos = input(f"Quantos anos você tem, {nome}?")
type(anos)
````

Como vimos lá atrás, você pode coagir essa entrada para outro formato usando funções como `int()` e `float()`:

````python
anos = int(input(f"Quantos anos você tem, {nome}?"))
type(anos)
````

Como o objeto criado de `input()` é do tipo textual, você pode usar os métodos habituais:

````python
email = input("Qual é o seu email?").upper()
email
````

## Usando a biblioteca `friendly`

É muito frequente encontrar erros quando fazemos código. `Python` é uma liguagem em que os avisos de erro são muito úteis, se comparada com outras linguagens (sim, `R` é  de você que eu estou falando kkk).

No entanto, mesmo para um iniciante, os erros podem ser um pouco confusos. Por isso, há uma biblioteca chamada `friendly` ("amigável"), que retorna sugestões mais compreensíveis nos avisos de erro. Infelizmente, os avisos de erro só estão disponíveis em inglês, mas ainda assim ajuda muito.

Para instalar, há três opções principais:

1. Digite `pip install friendly` no prompt do Anaconda, ou
2. Digite `python -m pip install friendly` no prompt do Windows, ou
