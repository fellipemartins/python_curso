<a name="ecossistema">
    <H1>Ecossistema de ciência de dados em Python</H1>
</a>

De forma genérica o ecossistema de análise de dados (que faz parte de um fluxo maior de ciência de dados) em `Python` inclui a própria linguagem e diversos pacotes / módulos. Conforme o crescimento do número de pacotes disponíveis, houve a necessidade de gerenciar esse portfólio de pacotes de forma ordenada. Surgiram então dois sistemas comparáveis, o `pip` e o `conda`.


## Gerenciadores de pacotes `pip` e `conda`
O `pip` é a ferramenta recomendada pela `Python Packaging Authority` para baixar e usar pacotes direto do repositório Python Packege Index (`PyPI`). O `conda`, por outro lado, abarca pacotes de diversas linguagens incluindo `Python`, gerenciados pelo repositório `Anaconda` e `Anaconda Cloud`. Enquanto as bibliotecas `pip` são compiladas localmente, as `conda` já vêm pré-compiladas. De forma simplificada, ambos instalam pacotes, mas `conda` pode conter material em outras linguagens.


## Anaconda distribution
Acredito que a maioria dos inciantes em análise de dados no ecossistema `Python` usem a [distribuição Anaconda](https://www.anaconda.com/products/individual). Isso é natural porque facilita muito a vida de quem começa ao fornecer num só procedimento as ferramentas iniciais básicas. Isso inclui acesso a centenas de pacotes já incluídos, e as IDEs `Spyder`(*scientific python development*) e `Jupyter`. A distribuição Anaconda também se integra muito bem com outras ferramentas (como o Orange ou RStudio):

<a name="alternativas">
    <H1>Alternativas ao Anaconda</H1>
</a>

## `Miniconda`

A distribuição `Anaconda` é provavelmente a mais usada em ciência de dados com `Python`. Mas, trata-se de uma distribuição pesada, que já vem com muita coisa útil que, no entanto, é usada menos frequentemente. Isso ocupa um espaço desnecessário no computador, especialmente quando não se tem uma máquina de última geração.
Uma opção alternativa menos pesada é a distribuição `Miniconda` ([link aqui](https://docs.conda.io/en/latest/miniconda.html)) que pode ser uma solução. Se for o seu caso, pode ser uma boa! A `Miniconda` é o cerne da Anaconda, mas você precisa instalar manualmente o que você precisar.


## `Miniforge`
Outra alternativa, para fugir de questões de licença de uso empresarial da Anaconda, é usar a distribuição `Miniforge` ([link aqui](https://github.com/conda-forge/miniforge)). Essa distribuição usa o mesmo sistema de gestão `conda`, mas sem as amarras legais.


## Google `Colab`
Você pode também usar o Google Colaboratory (`Colab`) na nuvem ao invés de ter uma distribuição instalada. Como todo serviço em nuvem, há as vantagens e desvantagens habituais (disponibilidade do serviço, espaço, etc). Uma desvantagem a ser considerada é que alguns pacotes não dão suporte ao `Colab`. Por exemplo, uma das bibliotecas que vamos utilizar na disciplina não funciona no `Colab`. Isso não impede, claro, que você use o `Colab`, mas alguns recursos podem ficar de fora.


## Notebooks no Kaggle
Por último, você pode utilizar os notebooks direto dentro da plataforma `Kaggle` ([link aqui](https://www.kaggle.com/)).

    <H1>Bibliotecas <code>Python</code> principais usadas no curso</H1>
</a>

Iremos utilizar diversas bibliotecas ao longo do curso e os alunos devem baixar e instalar as mesmas antes da próxima aula. A maioria delas é pequena e leva pouco tempo para baixar e instalar em banda larga. A maioria já vem instalada por padrão com o `Anaconda`, mas algumas precisam ser instaladas.

Abra o `Anaconda Prompt` e digite os comandos abaixo (um por vez):

```
pip install numpy
pip install pandas
pip install unidecode
pip install pyjanitor
pip install pandas_flavor
pip install matplotlib
pip install seaborn
pip install plotly
```

Você pode usar `pip install` e colocar as bibliotecas uma após a outra, separadas por espaço, mas é legal instalar uma a uma e testar seu funcionamento.
