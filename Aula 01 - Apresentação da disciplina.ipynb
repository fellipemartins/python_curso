<a name="avaliacao">
<H1>Avaliação</H1>
</a>

O sistema de avaliação na disciplina de Análise de dados para IoT será composto de:

- Projeto a ser apresentado ao fim da disciplina (80%)
    - Organização
    - Utilização de recursos
    - Lógica e apresentação
- Exercícios (10%)
- Participação em aula (10%)

A principal avaliação da disciplina consistirá na apresentação de um projeto prático de Análise de Dados. O projeto deve ser apresentado ao fim da disciplina e o o conteúdo do projeto disponibilizado para o professor e demais colegas (em arquivo *notebook* Juptyer, Google Colab, no GitHub, etc.).

Espera-se do projeto uma análise de dados utilizando-se do (mas não se limitando ao) conteúdo da disciplina . Dá-se preferência para exploração de dados específicos a IoT, mas não será penalizado caso o aluno não tenha acesso a dados reais de IoT - mais detalhes com o professor.

<a name="materiais">
<H1>Materiais</H1>
</a>

Vamos utilizar extensivamente a linguagem de programação `Python`. Também utilizaremos principalmente materiais compostos de *notebooks* `Jupyter` (isto que vocês veem na tela no momento).

Sugere-se, portanto, o *download* da distribuição `Anaconda` que já inclui a última versão do `Python` e das principais bibliotecas de análise de dados (`pandas`, `numpy`, etc.).

Iremos utilizar diversas bibliotecas ao longo do curso e os alunos devem baixar e instalar as mesmas antes da próxima aula. A maioria delas é pequena e leva pouco tempo para baixar e instalar em banda larga. A maioria já vem instalada por padrão com o `Anaconda`, mas algumas precisam ser instaladas.

Abra o `Anaconda Prompt` e digite os comandos abaixo (um por vez):

```
pip install numpy
pip install pandas
pip install unidecode
pip install pyjanitor
pip install pandas_flavor
pip install matplotlib
pip install seaborn
pip install plotly
```

# Alternativas ao Anaconda:

## Usando `Miniconda` ao invés de `Anaconda`

A distribuição `Anaconda` é provavelmente a mais usada em ciência de dados com `Python`. Mas, trata-se de uma distribuição pesada, que já vem com muita coisa útil que, no entanto, é usada menos frequentemente. Isso ocupa um espaço desnecessário no computador, especialmente quando não se tem uma máquina de última geração.

Uma opção alternativa menos pesada é a distribuição [`Miniconda`](https://docs.conda.io/en/latest/miniconda.html#:~:text=Miniconda%20is%20a%20free%20minimal,zlib%20and%20a%20few%20others.) que pode ser uma solução. Se for o seu caso, pode ser uma boa!
